## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Handling missing values in 'Customer_Location' (categorical)
auto_df['Customer_Location'].fillna(auto_df['Customer_Location'].mode()[0], inplace=True)
````

````python
# Handling missing values in 'Total_Past_Communications' (numerical)
auto_df['Total_Past_Communications'].fillna(auto_df['Total_Past_Communications'].mean(), inplace=True)
````

````python
# Handling missing values in 'Total_Links' (numerical)
auto_df['Total_Links'].fillna(auto_df['Total_Links'].mean(), inplace=True)
````

````python
# Handling missing values in 'Total_Images' (numerical)
auto_df['Total_Images'].fillna(auto_df['Total_Images'].mean(), inplace=True)
````

````python
# Use the isna() or isnull() method to identify missing values
missing_values = auto_df.isna()

# Count the missing values in each column
missing_count = missing_values.sum()
missing_count
````

### What all manipulations have you done and insights you found?

issing values in the 'Customer_Location,' 'Total_Past_Communications,' 'Total_Links,' and 'Total_Images' columns of the 'auto_df' dataframe were handled using appropriate strategies. For the 'Customer_Location' column, a categorical variable, the missing values were imputed with the mode of the column, representing the most frequently occurring location. In the case of numerical columns ('Total_Past_Communications,' 'Total_Links,' 'Total_Images'), missing values were filled with the mean of each respective column. This approach ensures that the missing values are replaced with representative values, preserving the overall distribution of the data. ### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.


The Email_Type of the campaign has no statistically significant effect on the Email_Status.

Null Hypothesis (H0): There is no association between Email_Type and Email_Status.

Alternative Hypothesis (H1): There is a statistically significant association between Email_Type and Email_Status.


#### 2. Perform an appropriate statistical test.

````python
# Importing necessary libraries
import pandas as pd
from scipy.stats import chi2_contingency

# Creating a contingency table (cross-tabulation) from the 'Email_Type' and 'Email_Status' columns in the 'auto_df' DataFrame
cross_tab = pd.crosstab(auto_df['Email_Type'], auto_df['Email_Status'])

# Performing the chi-square test on the contingency table
chi2, p, _, _ = chi2_contingency(cross_tab)

# Displaying the results of the chi-square test
print(f'Chi-square value: {chi2}')
print(f'P-value: {p}')

# Setting the significance level (e.g., 0.05)
alpha = 0.05

# Checking if the p-value is less than the significance level
if p < alpha:
    # Rejecting the null hypothesis: There is a significant association between Email_Type and Email_Status
    print("Reject the null hypothesis: There is a significant association between Email_Type and Email_Status.")
else:
    # Failing to reject the null hypothesis: There is no significant association between Email_Type and Email_Status
    print("Fail to reject the null hypothesis: There is no significant association between Email_Type and Email_Status.")



````

##### Which statistical test have you done to obtain P-Value?

The statistical test performed to obtain the p-value is the chi-square test of independence

##### Why did you choose the specific statistical test?

The chi-square test of independence was chosen for this analysis because it is appropriate for examining the association between two categorical variables, such as Email_Type and Email_Status. The test helps determine whether the observed distribution of data significantly differs from the distribution that would be expected if the variables were independent. In this case, it was used to assess whether there is a significant association between the type of email campaign (Email_Type) and the email status (Email_Status).

Feature selection is the process of choosing relevant features from a dataset to enhance model performance by reducing complexity, improving interpretability, and mitigating the risk of overfitting, leading to more efficient and accurate machine learning models.

````python
from sklearn.feature_selection import VarianceThreshold
import seaborn as sns
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
````

These libraries and functions collectively provide tools for analyzing and selecting features based on their variance, visualizing feature correlations, and detecting multicollinearity in the data.

````python
# Dropping Constant and Quasi Constant Feature
def dropping_constant(data):
    var_thres = VarianceThreshold(threshold=0.05)
    var_thres.fit(data)
    concol = [column for column in data.columns if column not in data.columns[var_thres.get_support()]]
    if "Email_Status" in concol:
        concol.remove("Email_Status")
    else:
        pass
    print(f'Columns dropped: {concol}')
    auto_df_changed = data.drop(concol, axis=1)
    return auto_df_changed
````

The purpose of this process is typically to reduce dimensionality by removing features that don't provide much information due to low variance. Low-variance features may not contribute significantly to the model's performance and can be considered as less informative.
````python
# Calculating VIF
def calc_vif(X):
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

# Checking Variable Inflation Factor
# the independent variables set
X_vif = auto_df_changed.copy()

# VIF dataframe
vif_auto_df = calc_vif(X_vif)
print(vif_auto_df)
````

By examining the VIF values, you can identify variables that might be highly correlated with others. Typically, VIF values above a certain threshold (e.g., 5 or 10) may indicate problematic levels of multicollinearity, and steps can be taken to address it, such as removing one of the correlated variables.

````python
# Check Feature Correlation and finding multicollinearity
def correlation(df, threshold):
    col_corr = set()
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return list(col_corr)

# Dropping highly correlated values
correlated_features = correlation(auto_df_changed, 0.6)
df_removed =auto_df_changed.drop(correlated_features, axis=1)
print("Dropped Features:", correlated_features)
print("New DataFrame Shape:", df_removed.shape)

# Checking VIF post-dropped features
# the independent variables set
X_vif_post_dropped =auto_df_changed.copy()

# VIF dataframe
vif_data_post_dropped = calc_vif(X_vif_post_dropped)
print(vif_data_post_dropped)
````

Overall, the code aims to address multicollinearity issues by identifying and removing highly correlated features from the dataset. By dropping correlated features, the remaining set of features is expected to have lower multicollinearity, as reflected in the updated VIF values. This helps in improving the stability and interpretability of regression models.

### 9. Handling Imbalanced Dataset

##### Do you think the dataset is imbalanced? Explain Why.

Yes, the dataset exhibits class imbalance, with a significant number of emails being ignored compared to those being read and acknowledged, as observed during exploratory data analysis. This imbalance can lead to biased classification favoring the majority class. To address this issue, two techniques, Random Undersampling and SMOTE (Synthetic Minority Oversampling Technique), are employed. Random Undersampling involves randomly selecting instances from the majority class to balance the dataset. On the other hand, SMOTE generates synthetic data for the minority class by selecting a point and creating synthetic points between the chosen point and its nearest neighbors. The goal is to mitigate the impact of class imbalance and enhance the performance of classification models, ultimately determining the most effective approach for the dataset.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score
import warnings

# Ignore UserWarnings
warnings.filterwarnings('always', category=UserWarning)

# Number of trees
n_estimators = [50, 80, 100]

# Maximum depth of trees
max_depth = [4, 6, 8]

# Hyperparameter Grid
param_dict = {'n_estimators': n_estimators,
              'max_depth': max_depth}

# Create an instance of the XGBClassifier
xgb_model = xgb.XGBClassifier()

# Grid search with 'roc_auc_ovr_weighted' as the scoring metric
xgb_grid = GridSearchCV(estimator=xgb_model,
                        param_grid=param_dict,
                        cv=5, verbose=2, scoring='roc_auc_ovr_weighted', error_score=0)

# Fit the Algorithm
xgb_grid.fit(X_train, y_train)

# Get the best model from the grid search
best_xgb_model = xgb_grid.best_estimator_

# Predict on the model
y_pred_optimized = best_xgb_model.predict(X_test)

# Get the predicted probabilities
test_probability_optimized = best_xgb_model.predict_proba(X_test)

# Print the best hyperparameters
print("Best Hyperparameters:", xgb_grid.best_params_)

# Print classification report for optimized model
print("Classification Report on Test Data (Optimized Model):")
print(classification_report(y_test, y_pred_optimized, target_names=['Ignored', 'Opened', 'Acknowledged']))

# Print ROC AUC score for optimized model
roc_auc_score_optimized = roc_auc_score(y_test, test_probability_optimized, multi_class='ovr', average='weighted')
print("ROC AUC Score on Test Data (Optimized Model):", roc_auc_score_optimized)
````
